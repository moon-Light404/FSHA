```python
n = 10
X = torch.from_numpy(getImagesDS(xpriv, n)).cuda() # 转成tensor
# print(X.shape)    # torch.Size([10, 1, 32, 32])
# x = getImagesDS(xpriv, n)
# print(x.shape)    # (10, 1, 32, 32) numpy

X = X.detach().cpu().permute(0,2,3,1).numpy()


def plot(X):
    n = len(X)
    X = (X+1)/2
    fig, ax = plt.subplots(1, n, figsize=(n*3,3))
    plt.axis('off')
    plt.subplots_adjust(wspace=0, hspace=-.05) # 图像之间紧密排列，减少空白区域。
    for i in range(n):
        ax[i].imshow((X[i]), cmap='Greys');  
        ax[i].set(xticks=[], yticks=[])
        ax[i].set_aspect('equal')
        
    return fig

fig = plot(X)
```
